```python
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
yrbss = pd.read_csv('../data/yrbss_2017.csv')
print(yrbss.shape)
```

```python
us_df = yrbss[yrbss['LocationAbbr'] == 'XX']
print(us_df.shape)
```

```python
Gender = ['Male', 'Female']
Race = ['Asian', 'Black or African American', 'Hispanic or Latino', 'White', 
        'Multiple Race', 'American Indian or Alaska Native', 'Native Hawaiian or Other Pacific Islander']
Sexuality = ['Gay or lesbian', 'Bisexual', 'Not sure', 'Heterosexual (straight)']
```

```python
group_vals = []
for risk_question in us_df['Greater_Risk_Question'].unique():
    #print(risk_question)    
    tot_df = us_df[(us_df['Greater_Risk_Question'] == risk_question) & 
                (us_df['Grade'] == 'Total') & (us_df['Sex'] == 'Total') & 
                (us_df['Race'] == 'Total') & (us_df['SexualIdentity'] == 'Total') & 
                (us_df['SexOfSexualContacts'] == 'Total')][['Greater_Risk_Data_Value']]
    if len(tot_df):
        tot_val = tot_df.iat[0,0]
        for gender in Gender:
            for race in Race:
                for sexuality in Sexuality:
                    member_val = us_df[(us_df['Greater_Risk_Question'] == risk_question) & 
                        (us_df['Grade'] == 'Total') & (us_df['Sex'] == gender) & 
                        (us_df['Race'] == race) & (us_df['SexualIdentity'] == sexuality) & 
                        (us_df['SexOfSexualContacts'] == 'Total')][['Greater_Risk_Data_Value']].iat[0,0]
                    group_vals.append([risk_question, gender, race, sexuality, member_val - tot_val])
vals_df = pd.DataFrame(group_vals)
```

```python
vals_df.to_csv('../data/vals_df.csv')
```
