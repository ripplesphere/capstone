```python
import pandas as pd
```

```python
yrbss = pd.read_csv('../data/yrbss_2017.csv')
print(yrbss.shape)
```

```python
question_results_df = pd.read_csv('../data/question_results_df.csv')
print(question_results_df.shape)
```

```python
us_df = yrbss[yrbss['LocationAbbr'] == 'XX']
print(us_df.shape)
```

```python
us_df[(us_df['Greater_Risk_Question'] == question_results_df.head(1)['question'][0]) & 
                (us_df['Grade'] == 'Total') & (us_df['Sex'] == question_results_df.head(1)['group'][0]) & 
                (us_df['Race'] == 'Total') & (us_df['SexualIdentity'] == 'Total') & 
                (us_df['SexOfSexualContacts'] == 'Total')][['Greater_Risk_Data_Value', 'Greater_Risk_Low_Confidence_Limit', 'Greater_Risk_High_Confidence_Limit']]
```

```python
val_diff = []
for index, row in question_results_df.iterrows(): 
        tmp_df = us_df[(us_df['Greater_Risk_Question'] == row['question']) & 
                (us_df['Grade'] == 'Total') & (us_df['Sex'] == 'Total') & 
                (us_df['Race'] == 'Total') & (us_df['SexualIdentity'] == 'Total') & 
                (us_df['SexOfSexualContacts'] == 'Total')][['Greater_Risk_Data_Value']] 
        val_diff.append(row['data_val'] - tmp_df.iat[0,0])
question_results_df['val_diff'] = val_diff
```

```python
question_results_df.sort_values(by=['val_diff'], ascending=False)
```
