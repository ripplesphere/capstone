```python
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
yrbss = pd.read_csv('../data/yrbss_2017.csv')
print(yrbss.shape)
us_df = yrbss[yrbss['LocationAbbr'] == 'XX']
print(us_df.shape)
```

```python
question_tots = []
for topic in us_df['Topic'].value_counts().items():
    topic_df = us_df[us_df['Topic'] == topic[0]]
    for subtopic in topic_df['Subtopic'].value_counts().items():
        subtopic_df = topic_df[topic_df['Subtopic'] == subtopic[0]]
        for risk_question in subtopic_df['Greater_Risk_Question'].value_counts().items():
            q_total = subtopic_df[(subtopic_df['Greater_Risk_Question'] == risk_question[0]) & 
                (subtopic_df['Sex'] == 'Female') &
                (subtopic_df['Grade'] == 'Total') &
                (subtopic_df['Race'] == 'Total') & (subtopic_df['SexualIdentity'] == 'Total') & 
                (subtopic_df['SexOfSexualContacts'] == 'Total')]
            q_val = q_total['Greater_Risk_Data_Value'].iloc[0]
            q_low = q_total['Greater_Risk_Low_Confidence_Limit'].iloc[0]
            q_high = q_total['Greater_Risk_High_Confidence_Limit'].iloc[0]
            question_tots.append([topic[0], subtopic[0], risk_question[0], q_low, q_val, q_high])
```

```python
tots_df = pd.DataFrame(question_tots)
```

```python
tots_df.to_csv('../data/female_tots_df.csv')
```
